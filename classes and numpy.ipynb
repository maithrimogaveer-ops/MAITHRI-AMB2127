```python
class crop:
  def __init__(self,name,season,expected_yield):
    self.name=name
    self.season=season
    self.expected_yield=expected_yield

  def total_production(self,area):
    return self.expected_yield*area

area=10
rice=crop("rice","kharif",5.2)
print("total rice production:", rice.total_production(area),"tones")
```

```python
class crop:
  def __init__(self,name,season,expected_yield):
    self.name=name
    self.season=season
    self.expected_yield=expected_yield

  def total_production(self,area):
    return self.expected_yield*area

rice=crop("rice","khari",5.2)
wheat=crop("wheat","rabi",4.1)
corn=crop("corn","kharif",6.8)
cotton=crop("cotton","kharif",22.3)
all_crops=[rice,wheat,corn,cotton] # Renamed list for clarity
for current_crop in all_crops: # Corrected variable name and iterated list
  production=current_crop.total_production(10) # Corrected method call
  print(f"{current_crop.name}:{production}tons")
```
